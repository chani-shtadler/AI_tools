```python
import os
from PIL import Image

def convert_images_to_grayscale(folder_path):
    """
    ×¢×•×‘×¨ ×¢×œ ×›×œ ×§×•×‘×¦×™ ×”×ª××•× ×” ×‘×ª×™×§×™×™×” × ×ª×•× ×”, ×××™×¨ ××•×ª× ×œ×©×—×•×¨-×œ×‘×Ÿ
    ×•×›×•×ª×‘ ××•×ª× ××—×“×© ×‘××•×ª×” ×ª×™×§×™×™×” (×“×•×¨×¡ ××ª ×”××§×•×¨).
    
    :param folder_path: ×”× ×ª×™×‘ ×”××œ× ×œ×ª×™×§×™×™×” ×”××›×™×œ×” ××ª ×”×ª××•× ×•×ª.
    """
    
    FOLDER_PATH = folder_path 
    
    # ×¡×•×’×™ ×§×‘×¦×™× ×©×”×§×•×“ ×™× ×¡×” ×œ×¢×‘×“:
    IMAGE_EXTENSIONS = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')
    
    print(f"âœ… ××ª×—×™×œ ×¢×™×‘×•×“ ×‘×ª×™×§×™×™×”: {FOLDER_PATH}")
    
    # ×‘×•×“×§ ×× ×”×ª×™×§×™×™×” ×§×™×™××ª
    if not os.path.isdir(FOLDER_PATH):
        print(f"âŒ ×©×’×™××”: ×”× ×ª×™×‘ ××™× ×• ×ª×™×§×™×™×” ×—×•×§×™×ª ××• ××™× ×• ×§×™×™×: {FOLDER_PATH}")
        return

    # ×¢×•×‘×¨ ×‘×œ×•×œ××” ×¢×œ ×›×œ ×”×§×‘×¦×™× ×‘×ª×™×§×™×™×”
    for filename in os.listdir(FOLDER_PATH):
        if filename.lower().endswith(IMAGE_EXTENSIONS):
            full_path = os.path.join(FOLDER_PATH, filename)
            
            try:
                # ×¤×•×ª×— ××ª ×”×ª××•× ×”
                img = Image.open(full_path)
                print(f"   ğŸ“‚ ×¢×™×‘×•×“ ×§×•×‘×¥: {filename}...")
                
                # ×××™×¨ ×œ×©×—×•×¨-×œ×‘×Ÿ. ×”××¦×‘ 'L' ×”×•× ×’×•×•× ×™ ××¤×•×¨ (Grayscale).
                # ×–×•×”×™ ×”×¤×§×•×“×” ×©××‘×¦×¢×ª ××ª ×”×”××¨×”:
                grayscale_img = img.convert('L') 
                
                # ×©×•××¨ ××ª ×”×ª××•× ×” ×”××¢×•×‘×“×ª, ××—×œ×™×£ ××ª ×”×§×•×‘×¥ ×”××§×•×¨×™!
                grayscale_img.save(full_path)
                
                print(f"   â­ {filename} ×”×•××¨ ×•×©×•××© ×‘×”×¦×œ×—×”.")
            
            except Exception as e:
                print(f"   âŒ ×©×’×™××” ×‘×¢×™×‘×•×“ {filename}: {e}")

    print("âœ… ×¡×™×•× ×”×¢×™×‘×•×“. ×›×œ ×”×ª××•× ×•×ª ×”×—×•×§×™×•×ª ×¢×•×“×›× ×• ×œ×©×—×•×¨-×œ×‘×Ÿ.")



## ğŸ“ ×”×™×›×Ÿ ×œ××§× ××ª ×”× ×ª×™×‘ ×œ×ª×™×§×™×™×”



# ----------------------------------------------------
#  ×”×¤×¢×œ×ª ×”×¤×•× ×§×¦×™×”
#  *** ×”××™×§×•× ×©×‘×• ×¢×œ×™×š ×œ×©×™× ××ª ×”× ×ª×™×‘ ×”××œ× ×œ×ª×™×§×™×™×” ×©×œ×š! ***
# ----------------------------------------------------
my_folder_path = r'C:\Users\User\Desktop\Chani_Studies\×›×œ×™ ×‘×™× ×”\lesson_2_hw\images' # *** ×”×—×œ×£ ××ª ×”× ×ª×™×‘ ×”×–×” ***

convert_images_to_grayscale(my_folder_path)
```
